```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import OneHotEncoder
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
```

```python
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
data.head()
```

```python
# 1. Get the numerical variables from our dataset.
num_data = data.select_dtypes([np.float64,np.int64])
num_data
```

```python
# 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
for col in num_data.columns:
        sns.histplot(x=num_data[col])
        plt.title(col)
        plt.show()
```

```python
# Although heavily skewed, most variables seem to follow a normal distribution. The worst cases are the following: 'Income',
# 'Months Since Last Claim', 'Months Since Policy Inception'
```

```python
# 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

for col in  num_data.columns:
    if min(num_data[col]) > 0:
        num_data[col] = stats.boxcox(num_data[col])[0]
    else:
        num_data[col] = np.log(num_data[col])
```

```python
for col in num_data.columns:
        sns.histplot(x=num_data[col])
        plt.title(col)
        plt.show()
```

```python
# Using boxcox or logarithmic transformations improves the distributions for the continuous variables. For the discrete ones,
# it does not make sense to normalize and instead we should use the original values.
```
